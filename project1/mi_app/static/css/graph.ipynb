```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función a graficar
def f(z):
    return (2*z - 2)/(3*z +2)

def g(z):
    return (5*z)/(3*z + 2)

# Crear los datos para el eje x
x = np.linspace(-10, 10, 400)

# Crear los datos correspondientes para los ejes y
y_f = f(x)
y_g = g(x)
print(f(10))
print(g(10))
# Crear la gráfica
plt.plot(x, y_f, label=r'$f(x) = x^2$', color='blue')
plt.plot(x, y_g, label=r'$g(x) = \sin(x)$', color='red')

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.title('Gráfica de las funciones $f(x) = x^2$ y $g(x) = \sin(x)$')

# Añadir leyenda
plt.legend()

# Mostrar la cuadrícula
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
